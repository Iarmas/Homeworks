```python
 import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import pandas as pd
import datetime
 # Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite", echo=False)
```

```python
engine.execute('SELECT * FROM measurement order by date desc LIMIT 5').fetchall()
```

```python
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
engine.execute('SELECT * FROM station LIMIT 5').fetchall()
```

```python
inspector = inspect(engine)
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)

Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

```python
date_a=input('Vacation beginning date (YYYY-MM-DD): ')
days=input('How long your vacation is (3-15): ')
```

```python
print(date_a)
date_a = datetime.datetime.strptime(date_a, '%Y-%m-%d').date()
print(type(date_a))
print(date_a)  # printed in default formatting
from dateutil.relativedelta import relativedelta
date_b=date_a - datetime.timedelta(days=int(days))
date_b
```

```python
date_c=date_a - relativedelta(years=1)
date_c
```

```python
q=session.query(Measurement.date,Measurement.prcp).filter(Measurement.date.between (date_c,date_a)).all()
q
```

```python
#conn=engine.connect()
df_rain=pd.DataFrame(q)
df_rain.sort_values(by='date')
```

```python
df_rain.set_index('date').head()
```

```python
df_rain.plot('date', 'prcp')
plt.xlabel("Date")
plt.ylabel("Precipitation")
plt.title("Precipitation Analysis")
#plt.legend(["Precipitation"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_rain.describe()
```

```python
locations = session.query(Measurement).group_by(Measurement.station).count()
print(locations)
```

```python
station = session.query(Measurement.station, func.count(Measurement.tobs)).group_by(Measurement.station).\
               order_by(func.count(Measurement.tobs).desc()).all()

max_station = station[0][0]    
print("The Station with more observations was",max_station,"with",station[0][1],"weather observations.")
```

```python
temperature = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.station == max_station).\
    filter(Measurement.date > date_c).\
    order_by(Measurement.date).all()
temperature
```

```python
temp_df=pd.DataFrame(temperature)
plt.hist(temp_df['tobs'],12)
plt.xlabel("Recorded Temperature")
plt.ylabel("# Observations")
plt.title("Station Analysis for Station " + max_station)
plt.show()
```
